## EDA and Data Visualisation with Matplotlib

## 1) Import packages and files

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
%matplotlib inline
pd.set_option('display.max_rows', 500)
```

## 2) Total columns 

### 2.1) Add a column for the total amount of copper mined per country since 1970

```python
cu_mined['Total mined per country'] = cu_mined.sum(axis=1)
```

```python
cu_mined.head()
```

```python
cu_mined.to_csv('Cu_mined.csv', index=False) # save file with total column for later analysis
```

### 2.2) Create a DataFrame containing the total amount of copper mined per year since 1970
Using .iloc[R, C], select all rows and all columns from 1970 (index 1) onwards. This creates a series of summed values. Use .drop to drop the last row which is the totalled 'Total mined per Country'. Use Transpose function to change the series to a DataFrame.

```python
year_total = cu_mined.iloc[:, 1:].sum()
year_total
```

```python
year_total.drop(labels=['Total mined per country'], inplace=True)
```

```python
year_total
```

```python
year_total_df = DataFrame(year_total).T #Transpose the series to a dataframe
```

```python
year_total_df
```

To add the missing country column reindex the year_total_df DataFrame with the columns labels from the cu_mined DataFrame. Pandas relabels all columns with labels from cu_mined DataFrame and fills in the missing values of the additional 'Country' column as NaN.

```python
year_total_df = year_total_df.reindex(columns = cu_mined.columns)

#reindex added a NaN column for Total mined per country as this is in the cu_mined DataFrame, drop this column
year_total_df.drop('Total mined per country', axis =1, inplace=True)
```

```python
year_total_df
```

Replace the NaN value in the country column with the string 'All countries'

```python
year_total_df['Country'].fillna('All countries', inplace = True)
year_total_df
```

```python
year_total_df.to_csv('tot_Cu_mined_per_yr.csv', index=False) #save as csv for later analysis
```

## 3) Reshaping the Pandas DataFrame
The data in the DataFrame is in a human readable form, there is a row for each country and a separate column for each year. However, this format is not good for analysis as it is not easy to call variables for plotting or to use functions such as 'Groupby'. 

The best format to arrange a DataFrame in is Tidy Data. Tidy data are data sets that are arranged such that columns represent one variable and each row represents an individual observation. For the cu_mined DataFrame this means having a column for Country, Year, Tonnes mined per country, and Tonnes mined per year by country. 

```python
cu_mined.shape
```

## 3.1) Data melting
use pd.melt() to rearrange the DataFrame so columns become rows and rows become columns

```python
melt_cu_mined = pd.melt(frame = cu_mined, id_vars =['Country','Total mined per country'], 
                        var_name = 'Year', value_name = 'Tonnes mined per year by country')
```

```python
melt_cu_mined.head()
```

```python
melt_cu_mined.describe() # statistics are for the whole data not just by year.
```

```python
sort_melt_cu_mined = melt_cu_mined.sort_values(['Tonnes mined per year by country'], ascending = False)
sort_melt_cu_mined.head(5)
```

### 3.2) Total Copper mined per year globally
Reshape the year_total_df and to create a line plot showing the amount of copper mined per year since 1970.

```python
year_total_df
```

```python
year_total_df_melt = pd.melt(frame = year_total_df, id_vars =['Country'], var_name = 'Year', value_name = 'Globally tonnes Cu mined per year')
```

```python
year_total_df_melt.head()
```

## 4) Plotting the data

### MATLAB-style interface

```python
fig = plt.figure()
year_total_df_melt.plot('Year','Globally tonnes Cu mined per year', 
                        figsize=(10,5),color='green', fontsize=12) 
plt.ylabel('Tonnes Cu mined')
plt.legend(['Tonnes Cu mined globally per year'])
plt.title('Tonnes of Copper mined globally from 1970 to 2018')
plt.show()
```

## Object-oriented interface

```python
fig, ax = plt.subplots()
fig.set_size_inches([15,5])
ax.plot(year_total_df_melt['Year'], year_total_df_melt['Globally tonnes Cu mined per year'],'g--',
        label = 'Tonnes Cu mined per year globally')
ax.set_ylabel('Tonnes Cu (million)')
ax.set_title('Tonnes of Copper mined per year globally', fontsize=16)
ax.set_xticklabels(year_total_df_melt.Year, rotation=60, fontsize=12)
ax.set_yticklabels(['0','6','8','10','12','14','16','18','20','24'], fontsize=12)
plt.show()
```

### Matplotlib Boxplot method
The first boxplot is created using Matplotlib boxplot, this accepts an array or sequence of vectors. It is not straight forward to group data to plot multiple boxplots on the one plot. Pandas boxplot is a quicker method to create boxplots.

```python
fig, ax = plt.subplots(figsize=(6,8))
ax.boxplot([melt_cu_mined['Total mined per country'], melt_cu_mined['Tonnes mined per year by country']])
plt.show()
```

### Pandas boxplot method
Boxplot is called on the DataFrame and the columns are passed as arguments

```python
bp = melt_cu_mined.boxplot(['Total mined per country','Tonnes mined per year by country'],
          fontsize=12, widths=0.5, figsize=(6,8), color='green')
bp.set_ylabel('Tonnes of Copper (million)', fontsize=12)
bp.set_yticklabels([0,0,20,40,60,80,100,120,140])
plt.show()
```

### Boxplot for each country
Pandas boxplot allows you to plot values grouped by another column using the 'by' parameter.One boxplot is drawn per grouped value in the column.

```python
bp1 = melt_cu_mined.boxplot('Tonnes mined per year by country', by='Country', fontsize=12, widths=0.5, figsize=(17,6), rot=90, grid=False, color = 'green')
bp1.set_ylabel('Tonnes Cu mined (million)', fontsize=13)
bp1.set_xlabel('')
bp1.set_yticklabels([0,0,1,2,3,4,5,6])
bp1.get_figure().suptitle('')
bp1.set_title('Tonnes of Cu mined by country from 1970 to 2018', fontsize=14)
plt.show()
```

### Examing the top producers of copper in the world
The boxplot shows that a number of countries have a range of values for tonnes mined over the 1970-2018 period, indicating that the amount they produced changed over time. These countries also have median values that lie above the upper limits of the other boxplots. 

Sorting the countries by the median of tonnes each has mined over the period shows a break between the median values between 175k for South Africa and 250k Kazakhstan. This coincides with a break in total mined that separates countries into two groups. Those that have in total produced less than 11M tonnes of copper and those that have produced more than 11M tonnes of copper since 1970. 

We will take countries in the >11M tonnes group as being the top producers and check if they have contributed most of the copper mined in the world to date. 

#### Create a DataFrame of the top copper producers, those that have in total mined more than 11M tonnes

```python
Top_cu_mines = melt_cu_mined[melt_cu_mined['Total mined per country'] > 11000000]
```

```python
Top_cu_mines['Country'].unique()  
```

```python
Top_cu_mines.head(52)
```

#### Group the Top_cu_mines DataFrame by year and sum the total tonnes mined by the 13 countries per year
Grouping by year sets the DataFrame index to year. To plot the data it's easiest to reset the index to integers.

```python
Top_mines_summed_year = Top_cu_mines.groupby('Year').sum()
Top_mines_summed_year.reset_index(inplace=True)
```

```python
Top_mines_summed_year
```

The first column above shows that the 13 countries have between them mined over 494 million tonnes of copper since 1970. the second column gives the total mined each year by the group of 13 countries

### Lineplot
Plot the amount of copper mined per year by the group of top producers and the amount of copper mined per year by all copper producing countries. 

```python
fig, ax = plt.subplots()
fig.set_size_inches([15,5])

ax.plot(year_total_df_melt['Year'], year_total_df_melt['Globally tonnes Cu mined per year'],'g--')
ax.plot(Top_mines_summed_year['Year'],Top_mines_summed_year['Tonnes mined per year by country'],'b--')

ax.set_ylabel('Tonnes Cu (million)', fontsize = 12)
ax.set_title('Tonnes of Copper mined from 1970-2018', fontsize=16)
ax.set_xticklabels(year_total_df_melt.Year, rotation=60, fontsize=12)
ax.set_yticklabels(['0','6','8','10','12','14','16','18','20','24'], fontsize=12)
ax.legend(['Tonnes Cu mined globally', 'Tonnes Cu mined by top 13 producers'], fontsize=12)
plt.show()
```

### Barplot

### Plot production per year per country from 1970 - 2018
Create a plot with a line for each of the top 13 producers showing the amount of copper each has produced per year since 1970.There are a number of ways to make this plot, some offer more felxibility than others. 

#### Option 1: Set the DataFrame index to Year. Using .groupby() group the DataFrame by 'Country'. Then select the 'Tonnes mined per year by country' column to plot using .plot(). This will be plotted against the index which is the Year. 
Not straight forward to set the line colour or style of each line using this method. 

```python
Year_index_top_cu = Top_cu_mines.set_index('Year')
Year_index_top_cu.groupby('Country')['Tonnes mined per year by country'].plot(figsize=(15,6),legend=True)
plt.text(40,5400000,'Chile', fontsize=14)
plt.show()
```

#### Option 2: Pivot the Top_cu_mines table so that each country row becomes a column containing values for tonnes mined per year for that country. Call .plot() on the pivoted DataFrame to plot each of the columns as separate lines against the Year index.

```python
Top_cu_mines_pivot = Top_cu_mines.pivot_table(index=['Year'],columns='Country', values='Tonnes mined per year by country')
Top_cu_mines_pivot.head(2)
```

```python
Top_cu_mines_pivot.plot(figsize=(15,6),linestyle='--', linewidth=1)
plt.axis('tight')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Tonnes Cu mined', fontsize=12)
plt.title('Tonnes of copper mined by top producers from 1970-2018', fontsize=12)
plt.legend(frameon=True, ncol=2, fancybox=True, shadow=True, borderpad=0.5);
```
